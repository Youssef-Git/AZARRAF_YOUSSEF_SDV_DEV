```python
for length in mower.price:
    if length >= 0 and length <= 300:
        mower['price'].replace(length,"0-300", inplace=True)
    elif length > 300 and length <= 500:
        mower['price'].replace(length,"301-500", inplace=True)
    elif length > 501:
        mower['price'].replace(length,"501-++", inplace=True)
```
